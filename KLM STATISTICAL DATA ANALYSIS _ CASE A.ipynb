```python
%matplotlib inline
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols

df = pd.read_excel('//kl1015cy/home$/klm76343/Desktop/000 Thesis/01 KLM/01 Documents/02 EXCEL/010 Statistics_bewteen_model_cases.xlsx',
                   'Sheet1')
prestige_model = ols("EGTM ~ dEta_is_HPC_2 + dWc_HPC_2", data=df).fit()
print(prestige_model.summary())
```

```python
fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.influence_plot(prestige_model, alpha  = 0.05, ax = ax, criterion="cooks")
```

```python
fig, ax = plt.subplots(figsize=(12,8))
fig = sm.graphics.plot_partregress("EGTM", "dEta_is_HPC_2", ["dEta_is_HPC_2", "dWc_HPC_2"], data=df, ax=ax)
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(prestige_model, "dWc_HPC_2", fig=fig)
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(prestige_model, "dEta_is_HPC_2", fig=fig)
```

```python
%matplotlib inline
import pandas as pd
from sklearn import preprocessing
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('//kl1015cy/home$/klm76343/Desktop/000 Thesis/01 KLM/01 Documents/02 EXCEL/01 EGTM Data Base Plots and Analysis.xlsx',
                   'Factory New (A3)')

df2 = pd.read_excel('//kl1015cy/home$/klm76343/Desktop/000 Thesis/01 KLM/01 Documents/02 EXCEL/01 EGTM Data Base Plots and Analysis.xlsx',
                   'KLM&Clients')

grouped = df2.groupby('Shop Visit')

min_max_scaler = preprocessing.MaxAbsScaler()

EGTM_FN_Scaled = min_max_scaler.fit_transform(df['EGTM'])
EGTM_1_Scaled = min_max_scaler.fit_transform(grouped["EGTM"].get_group('1st'))
EGTM_2_Scaled = min_max_scaler.fit_transform(grouped["EGTM"].get_group('2nd'))

data = [EGTM_FN_Scaled,EGTM_1_Scaled, EGTM_2_Scaled]
plt.boxplot(data)
title = "Centered and scaled to unit variance"
plt.title(title)
plt.show()
```
