```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

df = pd.read_excel('//kl1015cy/home$/klm76343/Desktop/000 Thesis/01 KLM/01 Documents/02 EXCEL/010 Statistics_bewteen_model_cases.xlsx',
                   'CASE-A')

data = df['EGTM']

# Fit a normal distribution to the data:
mu, std = norm.fit(data)

# Plot the histogram.
plt.hist(data, bins=25, normed=True, alpha=0.6, color='g')

# Plot the PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)
title = "Fit results: mu = %.2f,  std = %.2f for EGTM" % (mu, std)
plt.title(title)
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import iqr

df = pd.read_excel('//kl1015cy/home$/klm76343/Desktop/000 Thesis/01 KLM/01 Documents/02 EXCEL/010 Statistics_bewteen_model_cases.xlsx',
                   'CASE-A')

data = df['EGTM']
var = np.var(data)
std = np.std(data)
IQR = iqr(data)

print "For EGTM:"
print "Variance = ", var
print "St.dev = ", std
print "IQR = ", IQR
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('//kl1015cy/home$/klm76343/Desktop/000 Thesis/01 KLM/01 Documents/02 EXCEL/010 Statistics_bewteen_model_cases.xlsx',
                   'CASE-A')
x = df['EGTM']

LPC_ETA = stats.pearsonr(x, df['dEta_is_c_LPC_2 [%]'])
LPC_Wc_c = stats.pearsonr(x, df['dWc_c_LPC_2 [%]'])
LPC_Wc_d = stats.pearsonr(x, df['dWc_d_LPC_2 [%]'])
HPC_ETA = stats.pearsonr(x, df['dEta_is_HPC_2 [%]'])
HPC_Wc = stats.pearsonr(x, df['dWc_HPC_2 [%]'])
HPT_ETA = stats.pearsonr(x, df['dEta_is_HPT_2 [%]'])
HPT_Wc = stats.pearsonr(x, df['dWc_HPT_2 [%]'])
LPT_ETA = stats.pearsonr(x, df['dEta_is_LPT_2 [%]'])
LPT_Wc = stats.pearsonr(x, df['dWc_LPT_2 [%]'])

print "EGTM vs AM_Results person number coorelation and p-value:"
print "ETA LPC: ",LPC_ETA
print "Wc_{core} LPC: ",LPC_Wc_c
print "Wc_{bypass} LPC: ",LPC_Wc_d
print "ETA HPC: ",HPC_ETA
print "Wc HPC: ",HPC_Wc
print "ETA HPT: ",HPT_ETA
print "Wc HPT: ",HPT_Wc
print "ETA LPT: ",LPT_ETA
print "Wc LPT: ",LPT_Wc
```
